## This is Your Eloquent ->where()

| Laravel | Pandas |
|---------|--------|
| `->select('name', 'email')` | `df[['name', 'email']]` |
| `->where('age', '>', 30)` | `df[df['age'] > 30]` |
| `->first()` | `df.iloc[0]` |
| `->skip(5)->take(10)` | `df.iloc[5:15]` |

```python
import pandas as pd

df = pd.DataFrame({
    'filter_id': ['F001', 'F002', 'F003', 'F004', 'F005', 'F006'],
    'tds_output': [42, 78, 120, 35, 95, 180],
    'flow_rate': [2.1, 1.5, 0.8, 2.3, 1.2, 0.5],
    'age_days': [60, 180, 320, 15, 240, 350],
    'region': ['North', 'South', 'North', 'East', 'West', 'South']
})
df
```

```python
# Select columns
print("Single column (Series):")
print(df['tds_output'])

print("\nMultiple columns (DataFrame):")
df[['filter_id', 'tds_output', 'flow_rate']]
```

```python
# .loc - select by LABEL (like ->where)
print("Row 0, all columns:")
print(df.loc[0])

print("\nRows 1-3, specific columns:")
df.loc[1:3, ['filter_id', 'tds_output']]
```

```python
# .iloc - select by POSITION (like ->skip()->take())
print("First row:", df.iloc[0].values)
print("\nRows 2-4:")
df.iloc[2:5]
```

```python
# FILTERING - the most useful operation!
# "Show me filters with TDS > 80" (like ->where('tds', '>', 80))
high_tds = df[df['tds_output'] > 80]
print("High TDS filters:")
high_tds
```

```python
# Multiple conditions: use & (AND), | (OR)
# Filters with high TDS AND old age
critical = df[(df['tds_output'] > 80) & (df['age_days'] > 200)]
print("Critical filters (high TDS + old):")
critical
```

```python
# .query() - SQL-like syntax (cleaner for complex filters)
result = df.query('tds_output > 80 and region == "South"')
print("Query result:")
result
```
